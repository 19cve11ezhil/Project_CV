```python
import cv2
from ultralytics import YOLO
import pytesseract
import pandas as pd
from datetime import datetime

# Load YOLO model
model = YOLO(r'C:\Users\LENOVO\Downloads\best\best.pt')

# Video capture
video_path = 'C:\\Users\\LENOVO\\Downloads\\text_detection\\1mins_trim.mp4'
cap = cv2.VideoCapture(video_path)

# Get video properties
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_video_path = 'Downloads\\processed_video.avi'
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Initialize variables for shop names and time frames
shop_names = []
time_frames = []

ret = True
frame_count = 0
while ret:
    ret, frame = cap.read()
    frame_count += 1
    if not ret:
        break

    # Perform object detection on the frame
    results = model(frame)

    for result in results:
        for pred in result.boxes.xyxy:
            # Extract bounding box coordinates
            xmin, ymin, xmax, ymax = map(int, pred)

            # Extract ROI based on bounding box coordinates
            roi = frame[int(ymin):int(ymax), int(xmin):int(xmax)]

            # Preprocess the ROI for better OCR accuracy
            gray_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
            #blurred_roi = cv2.GaussianBlur(gray_roi, (5, 5), 0)
            #_, thresh_roi = cv2.threshold(blurred_roi, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
            
            # Configure Tesseract OCR engine
            config = '--psm 7 --oem 3 --dpi 300'  # Set page segmentation mode, OCR engine mode, and DPI

            # Apply OCR to the preprocessed ROI
            recognized_text = pytesseract.image_to_string(gray_roi, config= config, lang='eng')

            # Post-process the recognized text (if needed)
            cleaned_text = re.sub(r'\W+', ' ', recognized_text)  # Remove non-word characters

            # Store the recognized text and corresponding frame timestamp
            time_frame = datetime.utcfromtimestamp(frame_count // fps).strftime('%H:%M:%S')
            shop_names.append(cleaned_text)
            time_frames.append(time_frame)

            # Draw bounding box and recognized text on the frame
            cv2.rectangle(frame, (int(xmin), int(ymin)), (int(xmax), int(ymax)), (0, 255, 0), 2)
            cv2.putText(frame, cleaned_text, (int(xmin), int(ymin) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write the processed frame to the output video file
    out.write(frame)

    # Display the processed frame
    # cv2.imshow('Frame', frame)

    # Check for 'q' key press to exit the loop
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Release the video capture and video writer objects
cap.release()
out.release()
cv2.destroyAllWindows()

# Create a DataFrame with shop names and time frames
df = pd.DataFrame({'Shop Name': shop_names, 'Time Frame': time_frames})

# Write the DataFrame to an Excel file
output_excel_path = 'Downloads\\shop_names.xlsx'
df.to_excel(output_excel_path, index=False)

print("Processed video saved at:", output_video_path)
print("Shop names saved to Excel file at:", output_excel_path)
```

```python
import pandas as pd
import numpy as np

def aggregate_shop_names(input_excel_path, output_excel_path):
    # Read the input Excel file
    df = pd.read_excel(input_excel_path)

    # Filtering out the non-string, non-numeric, and non-space values 
    df['Shop Name'] = df['Shop Name'].apply(lambda x: ''.join([i for i in str(x) if i.isalpha()]))

    # Group by the time frame and aggregate the shop names 
    aggregated_df = df.groupby('Time Frame')['Shop Name'].apply(lambda x: ''.join(x)).reset_index()

    # Write the aggregated DataFrame to a new Excel file
    aggregated_df.to_excel(output_excel_path, index=False)

    print("Aggregated shop names saved to Excel file at:", output_excel_path)

# Example usage
input_excel_path = 'Downloads\\shop_names.xlsx'
output_excel_path = 'Downloads\\shop_names_agg.xlsx'

aggregate_shop_names(input_excel_path, output_excel_path)
```

```python
import pandas as pd
import spacy

# Load the English language model for SpaCy
nlp = spacy.load('en_core_web_sm')

# Read data from Excel file
excel_file_path = r'Downloads\shop_names_agg.xlsx'
df = pd.read_excel(excel_file_path)

# Iterate over each row and apply NER
for index, row in df.iterrows():
    character_representation = row['Shop Name']
    
    # Apply NER to the character representation
    doc = nlp(character_representation)
    
    # Extract named entities and their labels
    named_entities = [(ent.text, ent.label_) for ent in doc.ents]
    
    # Filter and post-process the named entities to identify shop names
    shop_names = [entity[0] for entity in named_entities if entity[1] == 'ORG']  # Assuming 'ORG' refers to organizations/shops
    
    # Output the identified shop names
    print("Shop Names:", shop_names)
```
